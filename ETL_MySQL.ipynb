## ETL -- EXTRACT from CSV TRANSFORM(null , duplicates) LOAD to MySQL


```python
import pandas as pd 
from sqlalchemy import create_engine
import mysql.connector as mysql
```

```python
raw_data = pd.read_csv(r'randomdataETL.csv')
raw_data.head() 
len(raw_data) #1005 rows × 8 columns
```

## Removing DUPLICATE data

```python
raw_data.duplicated().sum()
dup =raw_data[raw_data.duplicated()==True] 
raw_data.drop_duplicates(keep='first',inplace=True)
```

## Missing data

```python
raw_data.isnull().sum()
```

```python
raw_data[raw_data['ip_address'].isnull()]
```

```python
raw_data[raw_data['card_number'].isnull()]
```

```python
raw_data['card_number'].fillna(0,inplace=True)
raw_data['ip_address'].fillna(0,inplace=True)
```

```python
raw_data.isnull().sum()
```

# Connect to Database 

```python
# Credentials
connection = mysql.connect(
    host = "127.0.0.1",
    user = 'root',
    password = '')
```

```python
cursor = connection.cursor()
```

# LOAD to MySQL

```python
cursor.execute('create database if not exists simple_ETL')
```

```python
cursor.execute('use simple_ETL;')
cursor.execute('create table clean_data(id int not null primary key auto_increment);')
```

```python
host = "127.0.0.1"
user = 'root'
password = ''
engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}/simple_ETL')
```

```python
raw_data.to_sql('clean_data', con = engine , if_exists='replace' , index = False )

engine.dispose()
connection.close()
```
